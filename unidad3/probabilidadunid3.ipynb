```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([0, 1, 2, 3, 4, 5])
P = np.array([0.10, 0.25, 0.30, 0.20, 0.10, 0.05])

print("Suma de probabilidades:", np.sum(P))


print("\nDistribución de probabilidad:")
for x, p in zip(X, P):
    print(f"P(X={x}) = {p}")


E_X = np.sum(X * P)
print("\nValor esperado E[X]:", E_X)

E_X2 = np.sum((X**2) * P)
Var_X = E_X2 - E_X**2
Std_X = np.sqrt(Var_X)
print("Varianza Var(X):", Var_X)
print("Desviación estándar σ_X:", Std_X)

F_X = np.cumsum(P)
print("\nFunción de distribución acumulada F(x):")
for x_val, f_val in zip(X, F_X):
    print(f"F({x_val}) = {f_val}")

plt.step(X, F_X, where='post', label='F(x) acumulada')
plt.scatter(X, F_X, color='red')
plt.title('Función de Distribución Acumulada')
plt.xlabel('Número de células anormales (X)')
plt.ylabel('F(X)')
plt.grid(True)
plt.legend()
plt.show()
```
